````python
import re

def count_tokens(text):
    pattern = r'''
        (?:https?://\S+|www\.\S+\.\w+) |          # URLs
        (?:\S+@\S+\.\S+) |                        # E-Mails
        (?:\b\d{1,3}(?:,\d{3})+(?:\.\d+)?\b) |    # Zahlen mit Tausendertrennzeichen
        (?:\b\d+\.\d+\b) |                        # Dezimalzahlen
        (?:\b\w+-\w+(?:-\w+)*\b) |                # Bindestrichwörter
        (?:\b\w+(?:n't|'ll|'ve|'re|'d|'m)\b) |    # Echte Kontraktionen (ohne 's)
        (?:\b(?:Mr|Mrs|Ms|Dr|Prof|Rev|Hon|Capt)\.\b) | # Abkürzungen
        (?:\w+['’]\w+) |                          # Possessivformen (z. B. Smith's)
        (?:[.,!?;:]) |                            # Satzzeichen
        (?:\S\w*)                                  # Normale Wörter
    '''

    tokens = re.findall(pattern, text, re.VERBOSE | re.IGNORECASE)
    return len(tokens)

# Test mit dem Beispieltext
text = """
Mr. Smith's state-of-the-art website https://example.com had 3,000.50 visitors.
Don't miss user@example.com's O'Neill contest!
"""
print(count_tokens(text))  # Ausgabe: 16
````

Hinweis: Dieser Tokenizer ist sehr einfach und berücksichtigt keine Sonderfälle oder komplexere Sprachstrukturen. Für realistischere Anwendungen werden oft fortschrittlichere Tokenizer verwendet, die beispielsweise Satzzeichen, Sonderzeichen und Abkürzungen besser verarbeiten können.


````python
# prompt: einfacher Tokeniser ohne re modul

def simple_tokenizer(text):
    tokens = []
    current_token = ""
    for char in text:
        if char.isalnum() or char == "'":
            current_token += char
        elif current_token:
            tokens.append(current_token.lower())
            current_token = ""
    if current_token:
        tokens.append(current_token.lower())
    return tokens

text = "This is a simple sentence."
tokens = simple_tokenizer(text)
tokens
````

def simple_tokenizer(text):: Hier wird die Funktion simple_tokenizer definiert, die einen Text (text) als Eingabe entgegennimmt.
tokens = []: Eine leere Liste tokens wird erstellt, um die gefundenen Token zu speichern.
current_token = "": Eine leere Zeichenkette current_token wird erstellt, um das aktuelle Token während der Verarbeitung zu speichern.
Schleife zur Tokenisierung:

for char in text:: Der Code iteriert über jedes Zeichen (char) im Eingabetext.
if char.isalnum() or char == "'":: Wenn das aktuelle Zeichen alphanumerisch ist (isalnum()) oder ein Apostroph ('), wird es an current_token angehängt. Dies sammelt die Zeichen eines Tokens.
elif current_token:: Wenn das aktuelle Zeichen kein alphanumerisches Zeichen oder Apostroph ist und current_token nicht leer ist (d. h., ein Token wurde gesammelt), wird current_token in Kleinbuchstaben umgewandelt und der Liste tokens hinzugefügt. Dann wird current_token zurückgesetzt, um das nächste Token zu sammeln.
Letztes Token:

if current_token:: Nach der Schleife wird überprüft, ob current_token noch Zeichen enthält (d. h., ein Token am Ende des Textes wurde noch nicht hinzugefügt). Wenn ja, wird es in Kleinbuchstaben umgewandelt und der Liste tokens hinzugefügt.
return tokens: Die Funktion gibt die Liste der gesammelten Token (tokens) zurück.

**3. Kernkonzepte und Methoden im NLP**

*   **Daten:**
    *   Absolut fundamental für modernes NLP.
    *   Verfügbarkeit hat sich stark verbessert (kein mühsames Abtippen mehr).
    *   Problem: Zugang zu Daten wird teilweise wieder schwieriger (z.B. Schließung von Social-Media-APIs).
*   **Methoden:**
    *   **Reguläre Ausdrücke (Regex):**
        *   Werkzeug zur Mustererkennung in Text.
        *   Anwendung: Definition und Extraktion von *Tokens* (Wörter, Satzzeichen etc.) -> **Tokenisierung**.
        *   Beispiele für Regex-Syntax und deren Bedeutung werden kurz wiederholt (?, *, +, ^, $, /). Es wird empfohlen, diese zu wiederholen.
        *   Herausforderungen bei der Tokenisierung: Abkürzungen, Bindestriche, URLs, E-Mails, Zahlen, Satzzeichen am Ende, Groß-/Kleinschreibung (z.B. Satzanfang vs. Eigennamen).
        *   Man kann Regex nutzen, um z.B. Groß-/Kleinschreibung zu ignorieren oder spezifische Muster wie Telefonnummern als ein Token zu erkennen.
    *   **Statistische Methoden:**
        *   Wörter zählen, Satzlängen messen.
        *   Textcharakterisierung (z.B. Komplexität).
        *   Autorenstilanalyse (z.B. Vokabularreichtum).
        *   **Type-Token Ratio (TTR):** Verhältnis von einzigartigen Wörtern (Types) zu Gesamtzahl der Wörter (Tokens).
            *   Beispiel: "Fischers Fritz fischt frische Fische" - viele Tokens, wenige Types -> niedrige TTR.
            *   Kann ein Maß für die sprachliche Komplexität/Varianz sein (nicht unbedingt Qualität!).
            *   Beispielanalyse: Reden von Trump vs. Clinton zeigten Unterschiede in der TTR.
    *   **Maschinelles Lernen (ML):**
        *   Überwachtes und Unüberwachtes Lernen.
        *   **Künstliche Neuronale Netze (KNN):**
            *   Als Modell schon alt (seit 60ern/90ern), aber erst durch heutige Rechenleistung und Datenmengen effektiv nutzbar geworden.
            *   Grund für den aktuellen Hype und die Leistungsfähigkeit moderner Systeme.
        *   Traditionelle ML-Verfahren (SVM, Regression, Entscheidungsbäume, Naive Bayes) sind weiterhin relevant, besonders wenn Daten knapp sind.
    *   **Morphologie (Wortstruktur):**
        *   Analyse des Aufbaus von Wörtern.
        *   Relevant für: Pluralbildung, Komposita (im Deutschen sehr produktiv, z.B. Donaudampfschifffahrtskapitän), Flexion (Deklination, Konjugation), Derivation (Wortbildung, z.B. gehen -> Gang).
        *   Affixe: Präfixe (vorne), Suffixe (hinten), Infixe (innen), Zirkumfixe (drumherum).
        *   **Normalisierung:** Reduktion von Wortformen auf eine Grundform.
            *   **Stemming:** Regelbasiertes Abschneiden von Endungen (z.B. Porter-Stemmer für Englisch). Schnell, aber Ergebnis ist oft kein echtes Wort und kann Fehler machen (z.B. "organization" -> "organ", "doing" -> "doe"). Funktioniert für morphologisch reiche Sprachen wie Deutsch schlecht.
            *   **Lemmatisierung:** Zurückführen auf die tatsächliche Grundform (Lemma) mithilfe von Wörterbüchern und morphologischer Analyse. Langsamer, aber genauer. Besser für Deutsch.
*   **Evaluation:** **(Sehr wichtig!)**
    *   Bewertung der Performanz von NLP-Methoden/Systemen.
    *   **Pipeline-Problem:** Fehler am Anfang einer Verarbeitungskette (z.B. Tokenizer) pflanzen sich fort und werden nicht besser. Ein Tokenizer mit 97% Genauigkeit bedeutet 3% Fehler, die nachfolgende Schritte beeinflussen.
    *   **Datenaufteilung:** Üblich sind Training-, Entwicklungs- (Dev) und Test-Sets.
    *   **Ground Truth / Gold Standard:** Referenzdaten (oft manuell annotiert), gegen die verglichen wird.
    *   **Metriken (Quantitativ):**
        *   **Confusion Matrix:** Zeigt True Positives, True Negatives, False Positives, False Negatives.
        *   **Accuracy:** Gesamtgenauigkeit.
        *   **Precision:** Anteil der korrekt positiven Vorhersagen an allen positiven Vorhersagen.
        *   **Recall:** Anteil der korrekt positiven Vorhersagen an allen tatsächlichen Positiven.
        *   **F-Score:** Harmonisches Mittel aus Precision und Recall.
        *   (Bekannt aus Information Retrieval).
    *   **Qualitative Evaluation / Fehleranalyse:**
        *   Reine Zahlen (Metriken) reichen nicht. Man muss verstehen, *warum* Fehler passieren.
        *   Analyse der Fehlklassifikationen (False Positives/Negatives).
        *   Mustererkennung in Fehlern ist entscheidend für Verbesserungen.
        *   Achtung: Auch die Referenzdaten (Ground Truth) können Fehler enthalten!
        *   Skepsis bei 100% Genauigkeit – oft ein Zeichen für ein Problem (z.B. Overfitting oder Fehler in der Evaluation).

28. ◑ Define a function percent(word, text) that calculates how often a given word occurs in a text, and expresses the result as a percentage.

````python
def percent(word, text):
    """
    Berechnet, wie oft ein gegebenes Wort in einem Text vorkommt und drückt das Ergebnis als Prozentsatz aus.

    Args:
      word: Das Wort, dessen Häufigkeit berechnet werden soll.
      text: Der Text, in dem nach dem Wort gesucht werden soll.

    Returns:
      Den Prozentsatz, zu dem das Wort im Text vorkommt.
    """
    try:
        frequency = text.count(word)  # Zähle die Vorkommen des Wortes
        total_words = len(text)       # Gesamtzahl der Wörter im Text
        percentage = (frequency / total_words) * 100
        return percentage
    except ZeroDivisionError:  # Vermeidet eine Division durch Null, falls der Text leer ist
        return 0
````

````python
text = "Dies ist ein Beispieltext. Das Wort 'ist' kommt zweimal vor."
word = "ist"

prozent = percent(word, text)

print(f"Das Wort '{word}' kommt zu {prozent:.2f}% im Text vor.")
````

Okay, hier ist eine Zusammenfassung der Vorlesung basierend auf dem Transkript ab 14:10 Uhr:

**Hauptthema: Der Viterbi-Algorithmus und das Training von Hidden Markov Models (HMMs) in der Spracherkennung**

1.  **Wiederholung und Einleitung zum Viterbi-Algorithmus (14:10 - 14:13):**
    *   Die Vorlesung knüpft an vorherige Diskussionen über HMMs und endliche Automaten an.
    *   Ziel ist es, die Wahrscheinlichkeit einer Beobachtungssequenz (akustische Features) zu ermitteln und gleichzeitig die wahrscheinlichste zugrundeliegende Zustandssequenz (Phoneme/Wörter) zu finden.
    *   Der Viterbi-Algorithmus wird als Methode vorgestellt, um diese beiden Aufgaben effizient ("zwei Fliegen mit einer Klappe") zu lösen.

2.  **Das "Netz" der Möglichkeiten (Trellis) (14:13 - 14:16):**
    *   Spracherkennung wird als Navigation durch ein großes Netz (Trellis) von möglichen Lauten (Zuständen) mit Übergangswahrscheinlichkeiten dargestellt.
    *   Es gibt Anfangswahrscheinlichkeiten für Laute (z.B. der glottale Verschlusslaut am Satzanfang im Deutschen).
    *   Die zeitliche Komponente ist wichtig (kurze vs. lange Vokale), was bedeutet, dass man mit einer gewissen Wahrscheinlichkeit in einem Zustand (Laut) verweilen kann.
    *   Die Aufgabe ist, aus den beobachteten akustischen Features (MFCCs, LPCs etc.) zu bestimmen, welche Features zu welchem Laut gehören und welche Wahrscheinlichkeit diese Zuordnung hat.

3.  **Verborgene Zustände und der Forward-Algorithmus (Wiederholung) (14:16 - 14:21):**
    *   Erinnerung daran, dass die Phoneme/Wörter die "verborgenen Zustände" sind, die aus den beobachtbaren Features abgeleitet werden müssen.
    *   Der Forward-Algorithmus (bereits besprochen) berechnet die Wahrscheinlichkeit einer Beobachtungssequenz, indem er über alle möglichen Pfade im Trellis summiert.
    *   Jede Zelle im Trellis repräsentiert die Wahrscheinlichkeit, zu einem bestimmten Zeitpunkt in einem bestimmten Zustand zu sein, gegeben der bisherigen Beobachtungen.
    *   Dieser Prozess ist rekursiv und läuft "vorwärts" durch das Netz.
    *   Es wird betont, dass dieses Trellis sehr groß sein kann, abhängig von der Anzahl der möglichen Laute.

4.  **Details des Forward-Algorithmus (14:21 - 14:34):**
    *   Bei jedem Schritt werden Pfadwahrscheinlichkeiten, Übergangswahrscheinlichkeiten und Beobachtungswahrscheinlichkeiten (Wahrscheinlichkeit eines Symbols gegeben eines Zustands) aktualisiert.
    *   Obwohl viele Zustände durchlaufen werden, ist für die Berechnung im Zeitpunkt `t` primär der Zustand im Zeitpunkt `t-1` relevant, da die vorherige Geschichte in dessen Wahrscheinlichkeit implizit enthalten ist.
    *   Das Ergebnis ist eine Matrix, die diese Wahrscheinlichkeiten repräsentiert.

5.  **Vom Forward-Algorithmus zum Viterbi-Algorithmus (14:34 - 14:37):**
    *   Der Forward-Algorithmus liefert die Wahrscheinlichkeiten der Pfade, aber nicht direkt die wahrscheinlichste Sequenz der verborgenen Zustände.
    *   Der Viterbi-Algorithmus löst genau dieses Problem: Er findet die optimale (wahrscheinlichste) Sequenz von verborgenen Zuständen (Wörtern/Phonemen) gegeben einer Beobachtungssequenz und eines HMMs.

6.  **Intuition und Funktionsweise des Viterbi-Algorithmus (14:37 - 14:42):**
    *   Der Viterbi-Algorithmus berechnet die Wahrscheinlichkeit der Beobachtungssequenz *gleichzeitig* mit der besten Zustandssequenz.
    *   Dies ist effizienter als ein naiver Ansatz, bei dem jeder Pfad einzeln ausgewertet werden müsste.
    *   Er nutzt ebenfalls vorherige Pfadwahrscheinlichkeiten, Übergangswahrscheinlichkeiten und Beobachtungswahrscheinlichkeiten.
    *   Ein Schlüsselmerkmal ist das **Backtracking**: Es werden "Lesezeichen" (Pointer) in der Matrix gesetzt. Wenn ein Pfad sich als unwahrscheinlich erweist, kann zu einem vorherigen Punkt zurückgegangen und ein alternativer, potenziell besserer Pfad verfolgt werden, ohne das gesamte Trellis neu berechnen zu müssen.

7.  **Training von HMMs (14:42 - 14:53 & 15:16 - 15:33):**
    *   **Ziel:** Die Modellparameter (Übergangs- und Emissionswahrscheinlichkeiten) so anzupassen, dass die Wahrscheinlichkeit der beobachteten (trainierten) Daten maximiert wird.
    *   Man beobachtet Features (MFCCs etc.) und will die Wörter/Phoneme lernen.
    *   Es geht darum, wie wahrscheinlich ein Input (akustische Features) durch ein bestimmtes Wort/Phonem generiert wurde.
    *   **Parameterabschätzung:**
        *   Übergangswahrscheinlichkeiten: Wie oft folgt Zustand A auf Zustand B? (z.B. `P(Phonem_j | Phonem_i)`). Dies wird aus annotierten Trainingsdaten (wo akustische Signale mit Phonemen verknüpft sind) geschätzt.
        *   Emissions-/Beobachtungswahrscheinlichkeiten: Wie oft wird ein bestimmtes akustisches Feature-Set (z.B. ein MFCC-Vektor) in einem bestimmten Zustand (Phonem) beobachtet? (z.B. `P(akustisches_Feature | Phonem_j)`).
    *   Die Häufigkeit von Übergängen und die Gesamtzahl der Übergänge von einem bestimmten Zustand fließen in die Berechnung ein.
    *   Das akustische Modell enthält die Information, welche Phoneme auf welche folgen und wie Varianten (z.B. unterschiedliche Aussprachen) durch verschiedene Sprecher abgedeckt werden.
    *   Manuelle Transkriptionen sind entscheidend, um diese Modelle zu trainieren, weshalb Unternehmen wie Amazon und Apple menschliche Transkribierer einsetzen.

8.  **Struktur und Komplexität von Sprachmodellen (14:44 - 14:54):**
    *   Die Verbindung von Phonem-Sequenzen (z.B. für das Wort "five") und deren Wahrscheinlichkeiten ähnelt der Struktur von (vollständig verbundenen) neuronalen Netzen. Dies erklärt, warum Fortschritte bei neuronalen Netzen schnell in der Spracherkennung Anwendung fanden.
    *   Der Viterbi-Algorithmus kann das Berechnungsnetz effizienter gestalten, indem bestimmte Pfade ignoriert werden.
    *   Die Modelle werden komplexer durch N-Gramme (z.B. Bigramme), die den Kontext berücksichtigen (z.B. "das Haus" gefolgt von "ist" ist wahrscheinlicher als "das Haus" gefolgt von "bellt").
    *   Die Kombination aus akustischem Modell (Wahrscheinlichkeit der Features gegeben der Wörter) und Sprachmodell (Wahrscheinlichkeit der Wortfolgen) ist entscheidend.
    *   Freie Spracherkennungssysteme (wie Alexa, Siri) haben sehr große und komplexe Modelle.

9.  **Domänenspezifische vs. Grammatik-basierte ASR (14:54 - 15:09):**
    *   **Domänenspezifische ASR:** Modelle sind kleiner und auf bestimmte Bereiche (z.B. Medizin, Jura) zugeschnitten. Die Erkennung ist oft besser, aber das Vokabular ist eingeschränkt.
    *   **Grammatik-basierte ASR:**
        *   Haben einen sehr kleinen "Footprint" und können lokal (z.B. auf einem Raspberry Pi) laufen.
        *   Funktionieren mit explizit definierten Regeln (Grammatiken), z.B. für Navigationsbefehle ("Navigiere zu [Straße] [Nummer]").
        *   Sind robuster gegenüber Akzenten und Lärm, da der Suchraum stark eingeschränkt ist.
        *   Können nützlich sein für Smart-Home-Anwendungen.
        *   Sind jedoch weniger flexibel als freie Spracherkennungssysteme.

**Kernaussagen und Ablauf:**

1.  **Einleitung und manuelle Zusammenfassung (14:25):**
    *   Die Vorlesung beginnt mit der Frage, wie Studierende bei wissenschaftlichen Arbeiten Literatur zusammenfassen (Schlüsselwörter suchen, filtern, relevante Informationen definieren).
    *   Die Definition "relevanter Information" ist kontextabhängig und stark subjektiv (abhängig vom Informationsbedürfnis und Vorwissen des Lesers). Selbst bei gleichem Text können verschiedene Personen unterschiedliche relevante Informationen extrahieren.

2.  **Definition und Typen von Zusammenfassungen (ca. 14:28):**
    *   Eine Zusammenfassung ist eine kondensierte Repräsentation eines Inputs, geeignet zur menschlichen Weiterverarbeitung.
    *   **Beispiele:** Wikipedia "Exzellente Artikel" (der erste Abschnitt fasst den Rest zusammen), kurze Filmsynopsen.
    *   **Unterscheidung nach Zweck:**
        *   **Informative Zusammenfassung:** Liefert die wichtigsten Informationen (z.B. Wikipedia-Einleitung).
        *   **Indikative Zusammenfassung:** Soll Interesse wecken, ohne Details zu verraten (z.B. Film-Teaser, Klappentext).
    *   **Unterscheidung nach Erstellungsmethode:**
        *   **Extraktive Zusammenfassung:** Besteht aus wörtlich übernommenen Sätzen/Teilen des Originaltextes.
        *   **Abstraktive Zusammenfassung:** Formuliert Inhalte in eigenen Worten neu (paraphrasiert).

3.  **Parameter und Qualitätskriterien für Zusammenfassungen (ca. 14:33):**
    *   **Parameter:** Kompressionsrate (in Wörtern/Prozent), Kohärenz (Stichpunkte vs. Fließtext), Art der Quelldaten (homogen/heterogen, mono-/multilingual), Position (z.B. Abstract vor einem Paper).
    *   **Allgemeiner Prozess:** Analyse der Quelle(n) -> Transformation -> Synthese zur Zusammenfassung, unter Berücksichtigung der Parameter.
    *   **Qualitätskriterien:** Vollständigkeit, Präzision, Objektivität, Kürze, Verständlichkeit.

4.  **Manuelle Zusammenfassung im Detail (ca. 14:38):**
    *   Manuelles Zusammenfassen ist aufwendig. Eine Studie von 1996 zeigte, dass selbst geübte Personen ca. 12-20 Minuten pro wissenschaftlichem Artikel benötigen (Querlesen, relevante Passagen finden, organisieren, schreiben, überprüfen).
    *   Methoden: Nicht lineares Lesen, Nutzung der Dokumentstruktur, Verbindung von Passagen durch "discourse relations".

5.  **Automatische Zusammenfassung und "Shared Tasks" (ca. 14:40):**
    *   Das Thema kam bereits in den 1950ern auf.
    *   **Shared Tasks** (wie DUC/NIST): Wettbewerbe, bei denen Systeme auf standardisierten Datensätzen verglichen werden.
        *   Begann ca. 2001 mit "Single Document Summarization" (oft zu einfach, da bei Nachrichtenartikeln die ersten Sätze meist reichen).
        *   Später kamen komplexere Aufgaben hinzu: Multi-Document Summarization, Update Summarization (neue Informationen zu bekannten Themen), Topic-focused Summarization, "Multilingual Summarization" (eher sprachunabhängige Systeme für monolinguale Dokumente in verschiedenen Sprachen).
    *   **Herausforderungen:** Erstellung von Referenzdaten ist schwierig wegen der Subjektivität. Übersetzungsprobleme bei multilingualen Ansätzen.

6.  **Methoden der automatischen Zusammenfassung (ca. 14:58):**
    *   **Baselines:**
        *   Die ersten N Wörter/Sätze (Lead-Baseline).
        *   Zufällige Auswahl von Sätzen (Random Baseline).
    *   **Historische Methoden:**
        *   **Luhn (1958):** Statistisch, basierend auf Worthäufigkeit und Distanz signifikanter Wörter (Problem: Stoppwörter, keine Digitalisierung damals).
        *   **Edmundson (1960er):** Regelbasiert, nutzte Features wie Schlüsselwörter, Position im Text (Titel, Überschriften), Cue-Phrasen.
    *   **Lexikalische Ketten:** Identifizieren wichtiger Konzepte durch semantische Beziehungen zwischen Wörtern (z.B. Auto - Werkstatt - Reparatur). Erlaubt es, thematische Stränge zu verfolgen und Redundanz zu vermeiden.
    *   **Maximal Marginal Relevance (MMR):** Stammt aus der Informationssuche; wählt Sätze, die relevant zur Anfrage, aber gleichzeitig divers zueinander sind, um Redundanz zu minimieren.
    *   **Machine Learning Ansätze:**
        *   TF-IDF als Feature zur Satzbewertung.
        *   Satzlängenbeschränkungen.
        *   Graph-basierte Methoden (ähnlich PageRank) zur Identifikation wichtiger Sätze.

7.  **Zusammenfassung von gesprochener Sprache (Speech Summarization) (ca. 15:25):**
    *   **Herausforderungen:**
        *   Verschiedene Sprachformen (Monologe, Dialoge, Meetings).
        *   Disfluencies (Ähms, Füllwörter), Unterbrechungen, Satzabbrüche.
        *   Sprecherüberlappungen (Cross-Talk).
        *   Fehler in der automatischen Spracherkennung (ASR).
        *   Definition der Extraktionseinheit (Satz vs. Turn/Äußerung).
        *   Fokus (ein Sprecher vs. alle).
    *   Anwendungsideen: Automatische Agenda-Erstellung für Meetings.

    *   **Aspekte der Analyse:** Identifizierung von Entitäten (worüber wird gesprochen?), Aspekten (welche Eigenschaft wird bewertet?), Meinungsäußerer.
    *   **Ausdrucksformen:** Direkte, indirekte Meinungen und Vergleiche.
