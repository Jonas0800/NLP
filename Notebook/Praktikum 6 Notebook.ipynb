# 🧪 ROUGE-Evaluation automatischer Zusammenfassungen

Dieses Notebook evaluiert automatisch generierte Zusammenfassungen mithilfe der **ROUGE-Metrik** – einem Standardverfahren zur objektiven Bewertung von Textzusammenfassungen.

Verglichen werden Zusammenfassungen von zwei Algorithmen (**LexRank** und **Luhn**) mit einer Referenzzusammenfassung. Zwei verschiedene ROUGE-Implementierungen kommen zum Einsatz:
- `rouge-score` (Google)
- `evaluate` (HuggingFace)

```python
from rouge_score import rouge_scorer
import evaluate
```

## 📄 Beispielzusammenfassungen

```python
reference = """The solar system includes the sun and objects that orbit it. Earth is the only planet known to support life."""

generated_lexrank = """The solar system has many objects orbiting the sun. Earth is the only planet with known life."""

generated_luhn = """The sun is at the center and provides light. Earth is the only planet that supports life."""
```

## 📊 ROUGE mit Google rouge-score

```python
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

print("🔷 LexRank:")
print(scorer.score(reference, generated_lexrank))

print("\n🔶 Luhn:")
print(scorer.score(reference, generated_luhn))
```

## 📊 ROUGE mit HuggingFace evaluate

```python
rouge_eval = evaluate.load("rouge")

print("🔷 LexRank:")
print(rouge_eval.compute(predictions=[generated_lexrank], references=[reference]))

print("\n🔶 Luhn:")
print(rouge_eval.compute(predictions=[generated_luhn], references=[reference]))
```

## ✍️ Beobachtungen zur ROUGE-Evaluation

| Kriterium         | LexRank                              | Luhn                                |
|-------------------|--------------------------------------|-------------------------------------|
| ROUGE-1           | gute Wortabdeckung                   | ebenfalls solide                    |
| ROUGE-2           | bessere Phrasenübereinstimmung       | teilweise schwächer                 |
| ROUGE-L           | gute Strukturähnlichkeit             | meist kürzer, weniger Matches       |

### ✴️ Vergleichbarkeit der Pakete
- `rouge-score` und `evaluate` liefern ähnliche, aber nicht identische Werte.
- Unterschiede entstehen durch Tokenisierung, Stemming und Berechnungsdetails.

### ✴️ Fazit
- ROUGE ist hilfreich für den automatischen Vergleich von Algorithmen.
- Die Bewertung deckt **Überschneidungen** ab, aber nicht **semantische Qualität**.
- Eine Kombination aus **quantitativer** und **qualitativer** Bewertung ist ideal.
