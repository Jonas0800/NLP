# 🧪 Evaluation automatischer Zusammenfassungen mit Sumy

In diesem Notebook werden verschiedene Algorithmen zur Textzusammenfassung getestet und qualitativ evaluiert. Dabei nutzen wir die Python-Bibliothek **Sumy** mit den Algorithmen **LexRank**, **Luhn** und **LSA** auf Texten unterschiedlicher Genres und Sprachen.

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lex_rank import LexRankSummarizer
from sumy.summarizers.luhn import LuhnSummarizer
from sumy.summarizers.lsa import LsaSummarizer
```

## 📄 Beispieltexte aus verschiedenen Genres und Sprachen

```python
texts = {
    "Wikipedia (EN)": """The solar system consists of the sun and all the objects that orbit it, including planets, moons, asteroids, comets, and meteoroids.
    The sun is the central and most massive object in the solar system, providing light and heat to its surrounding bodies.
    Among the planets, Earth is the only one known to support life.""",

    "News (EN)": """The government unveiled a new tax reform package today, aimed at providing relief to middle-income families.
    The policy will reduce tax rates and simplify deductions. Economists are debating the long-term impact.""",

    "Social Media (EN)": """Can't believe I missed the bus again 😩. Mondays are the worst. At least I got coffee ☕. #mondaymood""",

    "Science (EN)": """Recent breakthroughs in genetic engineering have enabled scientists to modify organisms at the molecular level.
    CRISPR-Cas9 technology has opened new avenues in medicine and agriculture, though ethical concerns remain.""",

    "Wikipedia (DE)": """Die Donau ist mit über 2.800 Kilometern der zweitlängste Fluss Europas. 
    Sie fließt durch zehn Länder und mündet schließlich ins Schwarze Meer. Die Donau ist ein wichtiger Verkehrsweg 
    und Lebensraum für viele Tierarten."""
}
```

## 🔁 Zusammenfassungsfunktion

```python
def summarize_text(text, lang="english", sentences=2):
    parser = PlaintextParser.from_string(text, Tokenizer(lang))
    print("\n🔷 LexRank:")
    for s in LexRankSummarizer()(parser.document, sentences):
        print("-", s)
    print("\n🔶 Luhn:")
    for s in LuhnSummarizer()(parser.document, sentences):
        print("-", s)
    print("\n🔸 LSA:")
    for s in LsaSummarizer()(parser.document, sentences):
        print("-", s)
```

## ▶️ Anwendung auf alle Genres

```python
for genre, txt in texts.items():
    print("\n" + "="*30 + f" {genre} " + "="*30)
    lang = "german" if "(DE)" in genre else "english"
    summarize_text(txt, lang=lang)
```

## ✍️ Qualitative Beobachtungen

| Genre           | Algorithmus | Pro                              | Contra                            | Geeignet für           |
|-----------------|-------------|-----------------------------------|-----------------------------------|------------------------|
| Wikipedia (EN)  | LexRank     | gute Struktur, zentrale Infos     | evtl. technische Sprache           | Sachtexte              |
| Wikipedia (DE)  | Luhn        | verständlich, faktenbasiert       | Begriffe wiederholt                | Enzyklopädie           |
| News (EN)       | LSA         | komprimiert, faktenorientiert     | Kontext verkürzt                   | Kurznachrichten        |
| Social Media    | LexRank     | zentrale Aussage extrahiert       | bei Emojis/Hashtags ungenau        | eher ungeeignet        |
| Science (EN)    | Luhn        | erkennt Schlüsselaussagen         | evtl. Redundanz                    | Abstracts, Paper       |

### 💡 Fazit

- LexRank zeigt Vorteile bei zusammenhängenden Texten mit logischem Aufbau.
- Luhn ist einfach, aber bei sachlichen Texten sehr effektiv.
- LSA liefert starke Ergebnisse bei komplexen, inhaltlich dichten Texten – weniger geeignet bei kurzen Beiträgen.

Sprachlich funktioniert Sumy auch mit **Deutsch**, solange die Tokenizer-Einstellungen passen (z. B. `"german"`).
