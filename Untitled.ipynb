# 2. Pattern definiert

```python
pattern = r'\[[^\]]*\]|\([^)]*\)|[A-Za-zÄÖÜäöüß0-9_]+|[^\w\s]'
```

**Erklärung**  
- In `basic_tokenizer` verwenden wir ein Regex, um verschiedene Token-Arten zu erkennen:
  - `\[[^\]]*\]`: Eckige Klammern mit beliebigem Inhalt (z.B. `[5]`)
  - `\([^)]*\)`: Runde Klammern mit beliebigem Inhalt (z.B. `(digitale Signatur)`)
  - `[A-Za-zÄÖÜäöüß0-9_]+`: Wörter oder Zahlen
  - `[^\w\s]`: Einzelne Symbole oder Satzzeichen (z.B. `,`, `.`, `!`)
- Dies ist unsere **Basis-Zerlegung**, die wir anschließend weiter verfeinern werden.


# 3. Pattern in der Funktion basic_tokenizer mithilfe von Regex angewand

```python
def basic_tokenizer(text):
    pattern = r'\[[^\]]*\]|\([^)]*\)|[A-Za-zÄÖÜäöüß0-9_]+|[^\w\s]'
    tokens = re.findall(pattern, text)
    return tokens
```

# 4. Exceptionhandling der Tokens

**Erklärung** 
- In `final_merge` werden aufeinanderfolgende Token die beide mit einem Großbuchstaben anfangen zu einem Token zusammengeführt, um Namen von Personen ("Jonas Krumeich") und Städten ("New York") auf ein Token abzubilden.

```python
def final_merge(tokens):

    merged_tokens = []

    i = 0
    while i < len(tokens):
        token = tokens[i]

        # Zwei großgeschriebene Wörter nacheinander => "New York"
        if token.istitle() and i + 1 < len(tokens) and tokens[i + 1].istitle():
            merged_tokens.append(token + " " + tokens[i + 1])
            i += 2  # Wir haben 2 Tokens zusammengeführt
            continue

        # Standardfall: Übernehme das aktuelle Token unverändert
        merged_tokens.append(token)
        i += 1

    return merged_tokens
```
